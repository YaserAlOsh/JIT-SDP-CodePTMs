```python
apache_sources_path = 'D://Projects/JITSoftwareDefect/apache sources/'
projects = []
import os
import glob 
import pickle
import pandas as pd

for projectname in glob.glob(apache_sources_path+'/*_tokenized.pkl'):
    with open(os.path.join(apache_sources_path,projectname),'rb') as pf:
        p = pickle.load(pf)
        projects.append(p)
```

```python
def map_added_removed(added_removed):
    adf, rmf = added_removed
    objs = []
    i = 0
    for adl in adf:
        objs.append({'added':adl,'removed':rmf[i]})
    return objs
def convertDataFrameforCC2Vec(df):        
    cc2vec_df = pd.DataFrame()
    cc2vec_df[0] = df['commit_id']
    cc2vec_df[1] = df['buggy'].astype(int)
    # Get first line of log message only
    cc2vec_df[2] = df['commit_message'].str.join(' ').str.split(pat='\n',n=1,expand=True)[0]
    cc2vec_df[3] = list(map(map_added_removed,zip(df['added_code'],df['removed_code'])))
    return cc2vec_df
```

```python
for p in projects:
    p_df = convertDataFrameforCC2Vec(p.total)
    p.cc2vecTable = (p_df[0],p_df[1],p_df[2],p_df[3])
```

```python
use_lowercase = True
def add_dict(dict,key):
    if use_lowercase:
        key = key.lower()
    if key in dict:
        dict[key]+=1
    else:
        dict[key]=1
def add_msg_words(dict,msg):
    for w in msg.split(' '):
        add_dict(dict,w)
def add_code_words(dict,c_codes):
    for adrm in c_codes:
        for al in adrm['added']:
            for aw in al.split(' '):
                add_dict(dict,aw)
        for rl in adrm['removed']:
            for rw in rl.split(' '):
                add_dict(dict,rw)
def createDictionaryFromCC2VecTable(cc2vecTable):
    msg_dict, code_dict = dict(),dict()
    cc2vecTable[2].apply(lambda m: add_msg_words(msg_dict,m))
    cc2vecTable[3].apply(lambda c: add_code_words(code_dict,c))
    return (msg_dict,code_dict)
```

```python
for p in projects:
    p.cc2vecdict = createDictionaryFromCC2VecTable(p.cc2vecTable)
```
